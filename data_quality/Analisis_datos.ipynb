Paso 1: Lectura de archivo de referencia

```python
ruta_datos = "D:/Datos/Limpios"
ruta_reporte ="D:/Datos/reporte/unificado"
ruta_referencia = f"{ruta_datos}/base_datos.Altavista"
df_referencia = pd.read_csv(ruta_referencia)
ruta_procesados="D:/Datos/Procesados"
```

Paso 2: ajuste_columnas

```python
columnas_originales = df_referencia.columns.values
```

```python
columnas_nuevas = {columna: columna.split(".")[-1] for columna in columnas_originales}
```

```python
columnas_nuevas
```

```python
df_referencia.rename(columns=columnas_nuevas,inplace=True)
```

```python
df_referencia
```

```python
def renombrar_df(df: pd.DataFrame):
    columnas_originales = df.columns.values
    columnas_nuevas = {columna: columna.split(".")[-1] for columna in columnas_originales}
    #print(columnas_nuevas)
    return df.rename(columns=columnas_nuevas)
    
```

```python
columnas_nuevas
```

```python
rutas_archivos = glob.glob(f"{ruta_datos}/*")
df_concatenado = pd.DataFrame(columns= columnas_nuevas.values())
for ruta in rutas_archivos:
    df_temporal = pd.read_csv(ruta)
    df_temporal = renombrar_df(df_temporal)
    #print(df_temporal.columns.values)
    df_concatenado = pd.concat([df_concatenado,df_temporal],axis=0)
```

```python
df_concatenado = df_concatenado.infer_objects()
df_concatenado["comuna"] = df_concatenado["comuna"].astype("category")
df_concatenado["sexo_"] = df_concatenado["sexo_"].astype("category")
```

```python
df_concatenado.dtypes
```

Paso 3: Crear reporte generico de datos

```python
reporte_calidad = create_report(df_concatenado)
reporte_calidad.save(ruta_reporte)
```

Paso 4: guardar en diferentes archivos en base a la comuna

```python
grupos = df_concatenado.groupby("comuna")
for grupo in grupos.groups:
    grupos.get_group(grupo).to_csv(f"{ruta_procesados}/{grupo}.csv" ,index=False)
```
